```python
import seaborn as sns
import pandas as pd 


sns.set(style="darkgrid")
bearsData = pd.read_csv("bears.csv", index_col=0)
squirrelsData = pd.read_csv("squirrels.csv", index_col=0)

sns.regplot(x="Number of Bears", y="Speed", data=bearsData);
```

```python
dataSets = pd.concat([bearsData, squirrelsData])
combine = pd.merge(bearsData, squirrelsData, left_on = "Student ID", right_on = "Student ID")
combine.head()
sns.lmplot(x="Number of Bears", y="Speed", hue='Bag of Squirrels', data=combine);
```
